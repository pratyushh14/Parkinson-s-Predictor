```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm 
from sklearn.metrics import accuracy_score
```

```python
parkinson_data = pd.read_csv("parkinsons.csv")
parkinson_data.head()
```

```python
parkinson_data.isnull().sum()
```

```python
parkinson_data.describe()
```

```python
parkinson_data['PPE'].mean()
```

```python
parkinson_data['status'].value_counts()
```

```python
parkinson_data.drop('name', axis = 1, inplace=True)
```

```python
# group the data based on target variable (status)
parkinson_data.groupby('status',).mean()
```

```python
X = parkinson_data.drop(['status'], axis=1)
Y = parkinson_data['status']
X.head()
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=2)
```

```python
X_train
```

```python
X_test
```

```python
# Data Standarisation meaning we want all the data in common range
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
pd.DataFrame(X_train)
```

```python
# model training
model = svm.SVC(kernel='linear')
```

```python
# training the model using X_train and Y_train and using this it will find the hyperplane
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
training_data_accuracy
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
testing_data_accuracy
```

```python
# predictive system
input_data = (162.568, 198.346, 77.63, 0.00502,	0.00003, 0.0028, 0.00253, 0.00841, 0.01791,	0.168, 0.00793,	0.01057, 0.01799, 0.0238, 0.0117, 25.678, 0.427785,	0.723797, -6.635729, 0.209866, 1.957961, 0.135242)


numpy_array = np.asarray(input_data)

input_data_reshaped = numpy_array.reshape(1, -1)

std_data = scaler.transform(input_data_reshaped)

prediction = model.predict(std_data)
```

```python
if (prediction[0] == 1):
    print("The person has Parkinson Disease")
else:
    print("The person does not have Parkinson Disease")
```
